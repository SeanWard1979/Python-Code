```python
# Create a subset

df = ords_prods_merged[:1000000]
```

```python
# Define a function

def price_label(row):

  if row['prices'] <= 5:
    return 'Low-range product'
  elif (row['prices'] > 5) and (row['prices'] <= 15):
    return 'Mid-range product'
  elif row['prices'] > 15:
    return 'High range product'
  else: return np.nan
```

```python
# Apply the function

df['price_range'] = df.apply([price_label], axis=1)
```

```python
df['price_range'].value_counts(dropna = False)
```

```python
## Dividing up hours of the day by how busy they are ##  
result = []

for value in ords_prods_merged["order_hour_of_day"]:
    if value == 9 or value == 10 or value == 11 or value == 12 or value == 13 or value == 14 or value == 15 or value == 16:
        result.append("Most orders")
    elif value == 7 or value == 8 or value == 17 or value == 18 or value == 19 or value == 20 or value == 21 or value == 22:
        result.append("Average orders")
    else:
        result.append("Fewest orders")
```

```python
## Question 4 Creation of the new column to show busiest times of day ##
ords_prods_merged['busiest_period_of_day'] = result
```
